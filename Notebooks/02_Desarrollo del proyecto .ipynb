## Calidad de Datos

```python
#Eliminar duplicados
df.drop_duplicates(inplace = True)
df
```

```python
#Renombrar columnas
df.rename(columns = {'Name':'Title','Year':'Publication Year','User Rating':'Rating',}, inplace = True)
df
```

```python
#Convertir los tipos de datos
df.info()
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df.info()
```

## Analisis Exploratorio (EDA)

### Analisis de popularidad de autor

```python
p_autor = df['Author'].value_counts()
p_autor
```

### Calificación media por género

```python
rating_media = df.groupby('Genre')['Rating'].mean()
rating_media
```

## Exportamos los resultados

```python
p_autor.head(10).to_csv('top_autores.csv')
rating_media.to_csv('rating_media_genero.csv')
```
